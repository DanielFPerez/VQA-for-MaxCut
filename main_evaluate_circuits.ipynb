## Plot difference between classical and quantum 

```python
from src.utils import load_pickle, save_pickle
from src.qaoa import create_qaoa_circuit 
from src.eval_cirq import *
```

```python
in_size = (15,4)
```

```python
# Load the result data for different depths
depths = [1, 2, 3]
file_paths = [f"./results/grad_p{elem}_10nodes.pkl" for elem in depths]
result_data_list = [load_pickle(file_path) for file_path in file_paths]

# Extract the graphs and optimized parameters
graphs = result_data_list[0]['graphs']
optimized_parameters_list = [result_data['optimized_parameters'] for result_data in result_data_list]

# Compare QAOA with classical algorithm
qaoa_results, classical_results = compare_qaoa_with_classical_multi_p(graphs, optimized_parameters_list, depths)
```

```python
# Plot the results
fig_multiple_p = plot_multiple_p_results(qaoa_results, classical_results, depths, in_size=in_size, title="Gradient-based QAOA vs. Classical Algorithm for 10-node graphs")
```

```python
fig_multiple_p.savefig("./plots/10node_grad_multi-p_vs_classical.pdf")
```

```python
# Normalize QAOA results to classical results
normalized_results, classical_results = normalize_qaoa_to_classical(graphs, optimized_parameters_list, depths)
```

```python
# Plot the normalized bar chart
fig_normalized = plot_normalized_bar_chart(normalized_results, depths, in_size=(6,4), title="Gradient-based Normalized Average MaxCut Values")
```

```python
fig_normalized.savefig("./plots/10node_grad_multi-p_vs_classical_bars.pdf")
```
