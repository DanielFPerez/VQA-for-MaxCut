```python
import click
import cirq
import numpy as np
import networkx as nx
import scipy.optimize
import time


from src.utils import *
from src.qaoa import create_qaoa_circuit
from main_qaoa import *
```

```python
DEPTHS = [1, 2, 3]
OPTIMIZER = ['gradient', 'non-gradient']
SUBSET = 10

for elem in [10, 20]:
    print(f"\n############# GRAPHS OF {elem} nodes")
    if elem == 20:
        SUBSET = 5
    graphs = load_pickle(f"./data/graphs-{elem}nodes_100elems.pkl")[0:SUBSET]
    for optimizer in OPTIMIZER:
        print(f"###### OPTIMIZER: {optimizer} #####")
        for p in DEPTHS:
            print(f"\t###### p: {p} #####")
            times = []
            optimized_parameters = []
            for i, graph in enumerate(graphs):
                if i%5 == 0:
                    print(f"\t\tOptimizing QAOA for graph {i + 1}/{len(graphs)}", flush=True)
                t_start = time.time()
                optimal_params = optimize_qaoa(graph, p, optimizer)
                optimized_parameters.append(optimal_params)
                t_end = time.time() - t_start
                times.append(t_end)
                if optimizer == 'non-gradient':
                    print(f"\t\tTime taken for graph {i}: {t_end:.2f} seconds", flush=True)
                #print(f"Optimal parameters for graph {i + 1}: {optimal_params}", flush=True)
            mean = np.mean(times)
            std = np.std(times)
            print(f"\n\t\tAvg. time: {mean}")
            print(f"\t\tStd-dev. time: {std}")
        
```
